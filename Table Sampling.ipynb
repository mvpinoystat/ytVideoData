## Resample testData with replacement

```python

#needs a dataset as input as well as the size of the output
def resampleTable(dataset,  n_size):
    nrows = dataset.shape[0] 
    
    #cretae a list of rows
    listOfRows = [i for i in range(nrows)]
    
    #sample the rows with replacement:
    randomRows = np.random.choice(a = listOfRows, size = n_size, replace = True)
    
    #create a list to put row data:
    rowList = []
    
    #populate the list:
    for i in randomRows:
        rowList.append(dataset.loc[i,])
    
    #change the list into a pandas dataframe:
    newTable = pd.DataFrame(rowList)
    
    return newTable
```

```python
table1 = resampleTable(dataset = testData, n_size = 10)
table1
```

```python
table2 = resampleTable(dataset = testData, n_size = 10)
table2
```
